```python
import pandas as pd
from csv import QUOTE_NONE
data_dir = "./data/glue_data/STS-B/"
def read_table(loc):
    return pd.read_csv(loc, sep='\t', index_col=0, quoting=QUOTE_NONE)
train = read_table(data_dir+"train.tsv")
dev = read_table(data_dir+"dev.tsv")
test = read_table(data_dir+"test.tsv")
```

## Evaluation
Pearson correlation coefficient

```python
import numpy as np
import scipy
from scipy.stats import pearsonr
truth_train_score = train['score'].values
case_num = truth_train_score.shape[0]
```

```python
pred_train_score = np.random.random((case_num))
result = pearsonr(truth_train_score, truth_train_score/10 + 1)[0]
result = pearsonr(truth_train_score, pred_train_score)[0]
result_100 = abs(result) * 100 # in GLUE
result_100
```
